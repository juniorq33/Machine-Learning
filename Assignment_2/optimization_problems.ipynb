```python
import mlrose_hiive as mlrose
import matplotlib.pyplot as plt
import numpy as np
import time
import pandas as pd
```

```python
random_hill_climbing_results = []
simulated_annealing_results =[]
genetic_alg_results = []
mimic_results =[]

for size in range(10,100,10):
    weights= np.random.uniform(low=0.1, high=1, size=(size,))
    values= np.random.uniform(low=1, high=size, size=(size,))
    fitness = mlrose.Knapsack(weights, values)
    problem_fit = mlrose.DiscreteOpt(length = size,fitness_fn = fitness, maximize = True, max_val = 2)
    
    st_time=time.time()
    random_hill_climbing_results_curve = mlrose.random_hill_climb(problem_fit, max_iters = size*100, curve=True)
    et_time=time.time()
    random_hill_climbing_results.append([size,et_time-st_time,random_hill_climbing_results_curve[2][-1,0],random_hill_climbing_results_curve[2][-1,1]])

    st_time=time.time()
    simulated_annealing_results_curve = mlrose.simulated_annealing(problem_fit, max_iters = size*100, curve=True)
    et_time=time.time()
    simulated_annealing_results.append([size,et_time-st_time,simulated_annealing_results_curve[2][-1,0],simulated_annealing_results_curve[2][-1,1]])
    
    st_time=time.time()
    genetic_alg_results_curve = mlrose.genetic_alg(problem_fit, pop_size = size*10, mutation_prob=0.5, max_iters = size*100, curve=True)
    et_time=time.time()
    genetic_alg_results.append([size,et_time-st_time,genetic_alg_results_curve[2][-1,0],genetic_alg_results_curve[2][-1,1]])

    st_time=time.time()
    mimic_results_curve = mlrose.mimic(problem_fit, pop_size = size*10, max_iters = size*100, curve=True)
    et_time=time.time()
    mimic_results.append([size, et_time-st_time, mimic_results_curve[2][-1,0], mimic_results_curve[2][-1,1]])
```

```python
charts_title_dict = {1: "Time vs Problem Size", 2: "Best Fitness vs Problem Size", 3: "Function Evaluations vs Problem Size"}
charts_ylabel_dict = {1: "Time (secs)", 2: "Best Fitness", 3: "Function Evaluations"}


for n in range (1,4):
    random_hill_climbing_dataframe = pd.DataFrame(random_hill_climbing_results)[n]
    simulated_annealing_dataframe = pd.DataFrame(simulated_annealing_results)[n]
    genetic_alg_dataframe = pd.DataFrame(genetic_alg_results)[n]
    mimic_dataframe = pd.DataFrame(mimic_results)[n]
    df = pd.concat([random_hill_climbing_dataframe, genetic_alg_dataframe, simulated_annealing_dataframe, mimic_dataframe], ignore_index=True, axis=1)
    df.index = range(10,100,10)
    df.columns = ["RHC","GA","SA","MIMIC"]
    df.plot(marker='o', ylabel = charts_ylabel_dict[n], xlabel="Problem Size", title = charts_title_dict[n], figsize=(10,5))
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.model_selection import learning_curve

file_data = pd.read_csv('data.csv')
file_data = file_data.drop(["id"], axis=1)
file_data = file_data.drop(["Unnamed: 32"], axis=1)
file_data.diagnosis = [1 if i == "M" else 0 for i in file_data.diagnosis]
features_best = ['concave points_worst', 'perimeter_worst', 'concave points_mean', 'radius_worst', 'perimeter_mean', 'area_worst', 'radius_mean', 'area_mean', 'concavity_mean',
                    'concavity_worst',  'compactness_mean', 'compactness_worst', 'radius_se', 'perimeter_se', 'area_se']

x = file_data.drop(["diagnosis"], axis = 1)
y = file_data.diagnosis.values
x_all_train = x[features_best]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1)
train_x_features = x_train[features_best]
test_x_features = x_test[features_best]
print("Train dataset shape: ", train_x_features.shape )
print("Test dataset shape: ", test_x_features.shape )


```

```python
def plt_lc(lines, name, title):
    plt.figure()
    for (label, x, y) in lines:
        plt.plot(x, y, label = label)
    plt.xlabel("Train Size")
    plt.ylabel("Accuracy")
    plt.title(title)
    plt.legend()
    plt.savefig(name + ".png")
    plt.clf()
```

```python
model = MLPClassifier(solver='lbfgs', hidden_layer_sizes=(20), random_state=1)

train_size_abs, train_scores, test_scores = learning_curve(model, train_x_features, y_train, cv=10, train_sizes=[0.2, 0.4, 0.6, 0.8])
train_size_arr = []
train_acc = []
test_acc = []
for train_size, cv_train_scores, cv_test_scores in zip(train_size_abs, train_scores, test_scores):
    train_size_arr.append(train_size)
    train_acc.append(cv_train_scores.mean())
    test_acc.append(cv_test_scores.mean())
    print("Assignemnt 1", train_size, cv_train_scores.mean(), cv_test_scores.mean())

plt_lc([("Train", train_size_arr, train_acc), ("Test", train_size_arr, test_acc)], "results", "Assignment 1 Learning Curve")
```
